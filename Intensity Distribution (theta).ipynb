```python
import numpy as np

from matplotlib import pyplot as plt
```

```python
def wave(x, i, d, w):
    result = 0
    
    sinofx = np.sin(x)
    fraction = (np.pi * sinofx * d)/ w
    cos = np.cos(fraction)
    result = (cos** 2) * i
    
      
    return result
```

```python
def slitWidth(x, w, a):
    result = 0
    
    sinofx = np.sin(x)
    num = np.pi * a * sinofx / w
    numerator = np.sin(num)
    fraction = numerator / num 
    result = fraction ** 2
    
    return result
    
```

```python
def intenDis(x, i, d, w, a):
    
    result = wave(x, i , d, w) * slitWidth(x, w, a)
    
    return result
```

```python
%matplotlib inline

x = np.linspace(-np.pi/3, np.pi/3, 3000 )
w = 4e-7
i = 1
d = 2e-6
a = 4e-7



plt.plot(np.tan(x),  intenDis(x, i, d, w, a) )

plt.xlabel('Distance (m)')
plt.ylabel('Intensity (V^2/m^2)')
```
